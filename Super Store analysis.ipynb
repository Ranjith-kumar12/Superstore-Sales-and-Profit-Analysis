```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Dataset
data=pd.read_excel("C:\\Users\\DELL\\Downloads\\Sample - Superstore.xls")
```

```python
# Convert to DataFrame
df=pd.DataFrame(data)
```

```python
# First few row
df.head()
```

```python
df.tail()
```

```python
# Data summery
df.info()
```

```python
df.describe()
```

```python
# Handle Missing values
df.dropna(inplace=True)
```

```python
# Data cleaning removing duplicates

df.drop_duplicates(inplace=True)
```

```python
# Extracting Month & Year from order date

df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Year"]=df["Order Date"].dt.year
df["Month"]=df["Order Date"].dt.month
```

```python
# Sales Analysis

sales_by_year=df.groupby("Year")["Sales"].sum()
print(sales_by_year)
```

```python
# Plot sales over year+

plt.figure(figsize=(10,5))
sns.lineplot(x=sales_by_year,y=sales_by_year.values,marker="o")
plt.title("Total Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
# Category wise sales

plt.figure(figsize=(10,5))
sns.barplot(x=df["Category"],y=df["Sales"],estimator=np.sum)
plt.title("Total Sales by Catgory")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=35)
plt.show()
```

```python
#profit analaysis

plt.figure(figsize=(10,5))
sns.histplot(df["Profit"],bins=30,kde=True)
plt.title("Profit Distribution")
plt.xlabel("Profit Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# selecting only numeric columns
numeric_df = df.select_dtypes(include=['number'])
```

```python
# Correlation Heat map

plt.figure(figsize=(10,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.xlabel("Feature Correlation Headmap")
plt.show()
```

```python
# 10 profitable product 
top_products=df.groupby("Product Name")["Profit"].sum().nlargest(10)
top_products
```

```python
# plot 10 profitable product 

plt.figure(figsize=(10,5))
sns.barplot(x=top_products.values,y=top_products.index,palette="viridis")
plt.title("10 Profitable Products")
plt.xlabel("Total Profit")
plt.ylabel("Product Name")
plt.show()
```
